```python
num_of_country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
num_of_country_uses_python
```

```python
python_df = pd.concat([country_res, num_of_country_uses_python], axis = 'columns', sort = False)
```

```python
python_df
```

```python
python_df = python_df.rename(columns={'Country': 'Number of responded','LanguageWorkedWith':'Number of people knows python'})
```

```python
python_df
```

```python
python_df['pct_knows_python'] = (python_df['Number of people knows python'] / python_df['Number of responded from each country'])* 100
```

```python
python_df
```

```python
python_df.sort_values(by='pct_knows_python', ascending = False)
```
